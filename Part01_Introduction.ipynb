# **Ở chương 1 này, chúng ta sẽ được giới thiệu đến các thuật toán sẽ được học trong cuốn sách cũng như là lý do mà tác giả muốn truyền tải trong cuốn sách này cuối cùng là cuốn sách này giành cho những ai và làm thế nào để tận dụng cuốn sách này một cách tốt nhất**

## **1.1. Lời mở đầu**

Trong phần này tác giả có nhắc đến việc tầm quan trọng của mô hình học máy trong đời sống của chúng ta, chúng được viết lên bởi nhiều thuật toán khác nhau. Như vậy câu hỏi đặt ra là làm sao để hiểu dược chúng. Đó cũng chính là lý do mà cuốn sách này ra đời (Cuốn sách này hướng về những người không quá am hiểu về toán học như (Đại số tuyến tính, Thống kê, ...))   
Cách tốt nhất để chúng ta có thể mô tả một thuật toán máy học là
1. Số liệu thực tế được sử dụng trong thuật toán
2. Có thể tái sử dụng các thuật toán máy học để huấn luyện và sau đó dự đoán
3. Cho một ví dụ dễ để hình dung chính xác một con số làm sao để có thể tạo ra được 1 đầu ra **(output)**
Như vậy, khi đã đọc qua cuốn sách này ta sẽ hiểu rõ về cách hoạt động của mỗi thuật toán (cụ thể có 10 thuật toán) có thể xây dựng một thuận toán máy học tuỳ theo ngôn ngữ mà chúng ta chọn

## **1.2. Mô tả thuật toán**

Mô tả sẽ theo 3 quy tắc:
1. Mọi thuật toán đề được biểu diễn cẩn thận
2. Thuật toán học từ dữ liệu huấn luyện **(training data)**
3. Thuật toán thực hiện dự đoán khi gặp 1 dữ liệu mới dựa trên mô hình thuật toán đã học

## **1.3. Cấu trúc cuốn sách này**

Gồm có 4 phần:
1. Nền tảng **(Background)** cho các thuật toán học máy
  * Phần này diễn tả các thuật toán học máy hoạt động cũng như thực hiện giải quyết một vấn đề cơ bản như thế nào?. Cũng như hiểu thêm về nguyên lý hoạt động của các thuật toán. Cụ thể như:
    * Thuật ngữ được sử dụng trong học máy **(Machine Learning)** khi chúng mô tả dữ liệu
    * Framework để hiểu rõ các vấn đề mà thuật toán học máy giải quyết
    * Sự khác biệt quan trọng giữa thuật toán tham số và phi tham số
    * Sự tương phản giữa các vấn đề của học giám sát **(supervised,)**, không giám sát **(unsupervised)** và bán giám sát **(semi-supervised)**
    * bias và độ lệch **(variance)** ảnh hưởng đến các phương thức
    * Khắc phục vấn đề quá khớp dữ liệu **(overfitting data)**
2. Thuật toán máy học tuyến tính **(Linear Algorithms)**
  * Phần này giới thiệu về cách hoạt động của các thuật toán tuyến tính. Cụ thể như sau:
    * Tối ưu hoá độ dốc **(Gradient descent optimization)** được xem như là trái tim của nhiều thuật toán tuyến tính
    * Hồi quy tuyến tính **(Linear regression)** để dự đoán một giá trị thực
    * Hồi quy Logistic **(Logistic regression)** để phân loại giá trị cho 2 loại **(categories)**
    * Phân tích tuyến tính **(Linear discriminant analysis)** cho từ 2 biến categories trở lên
3. Thuật toán máy học không tuyến tính **(Nonlinear Algorithms)**
  * Thuật toán được xây dựng dựa trên các thuật toán tuyến tính ít sử dụng các dữ liệu thật và có khả năng làm việc với các bài toán khác nhau
    * Phân loại **(Classification)** và tuyến tinh **(regression)** dùng cây quyết định **(Decision Tree)**
    * **Naive Bayes** sử dụng sác xuát để phân loại
    * **K-Nearest Neighbors (KNN)** chỉ duy nhất tệp dữ liệu
    * **Learning Vector Quantization** cùng với **KNN** làm giảm chiều của dữ liệu
    * **Support Vecto Machine (SVM)** thuật toán phổ biến nhất
3. Thuật toán bảo gồm cả tuyến tính và không **(Ensemble Algorithms)**
  *  Thuật toán dựa trên kểt quả dự đoán của nhiều mô hình để thực hiện dự đoán làm cho dữu liệu thêm chính các hơn
    * **Bagging and Random Forests**, **Boosting ensemble and the AdaBoost**
