### Pylot API và Object-Oriented API
* Pylot API -> Nhanh chóng
* OO API -> Phức tạp phải có đủ dữ liễu

```python
x = np.linspace(0, 10, 1000)
plt.plot(x, np.sin(x), color = 'blue', linestyle='dashed', label = 'sin(x)');
plt.plot(x, np.cos(x), color= 'red', label = 'cos(x)');
plt.xlabel('Biến x')
plt.ylabel('f(x)')
plt.title('Sin(x) and Cos(x)')
# plt.xlim([0,4])
# plt.ylim([-1.5, 1])
# plt.axis([x min, x max, y min, y max])
plt.axis('equal') # Chỉnh giới hạn cột x và cột y

plt.legend();
# plt.show(); # Chỉ áp dụng khi chạy Matplotlib trên Terminal
```

##### Object-Oriented API

```python
# 1. Object-Oriented API chuẩn bị dữ liệu
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Điều chỉnh biểu đồ
fig, ax = plt.subplots(figsize=(5,5)) # Điều chỉnh độ rộng 5 chiều cao 5 cho biểu đồ

# 3. Tạo biểu đồ
ax.plot(x, y);

ax.set(title = 'Biều đồ đơn giản', 
      xlabel = 'x-axis',
       ylabel = 'y-axis');
```

## 2. Nhừng loại đồ thị phổ biến trong Matplotlib
* `Line` : Biểu đồ đường
* `Scatter` : Đồ thị rời rạc, phân tán
* `Bar` : Đồ thị dạng cột
* `Histogram`: Biểu đồ cột
* `Subplots`: Nhiều đồ thị khác

#### `Line char` : Biểu đồ đường

```python
# Tạo mảng 
x  = np.linspace(0, 10, 100)
x[:5]
```

```python
# OO API plot một line chart
fig, ax = plt.subplots()
ax.plot(x, x**3); # Tạo plot đường với cột x = x, y = x^3 
```

#### `Scatter` Biểu đồ phân tán

```python
# Pylot API
plt.scatter(x, np.exp(x)); # y = e^x, x = x
```

```python
# Chuẩn bị data mới 
rng = np.random.RandomState(0)

a = rng.randn(100)
b = rng.randn(100)

sizes = 1000 * rng.rand(100)
colors = rng.rand(100)
fig, ax = plt.subplots()

img1 = ax.scatter(a, b, s = sizes, c = colors, cmap = 'viridis', alpha = 0.6);

fig.colorbar(img1) # Tạo phân tán
```

#### `Bar Chart` Biểu đồ cột
* Vertical : Cột ngang
* Horizontal : Cột dọc

```python
# Chuẩn bị dữ liệu 
soft_drink_prices = {"Coke" : 10, 
                    "Pepsi" : 12,
                    "Sprite" : 8}
fig, ax = plt.subplots()
ax.bar(soft_drink_prices.keys(), soft_drink_prices.values());
ax.set(title="Bach Hoa Xanh soft_drink_prices", ylabel = "Price ($)");
```

```python
fig, ax = plt.subplots()
ax.barh(list(soft_drink_prices.keys()), list(soft_drink_prices.values()));
```

#### `Histogram` : Thể hiện tần suất dạng cột

```python
# Chuẩn bị data
np.random.seed(42)
student_height = np.random.normal(170, 10, 250)
student_height[:10], len(student_height)
```

```python
# Pyplot API
plt.hist(student_height, bins=30);
```
